```python
# Generamos un df con la data faltante para ver si es que otra columna nos entrega información que nos falta - no obtenemos nada asique seguimos analizando los nan
df_nan = df['Item_Weight'].isna()
df[df_nan]
```

Podemos observar que los elementos de `'Item_Weight'` fluctúan entre 4.55 y 21.35, con una moda de 12.857645.

Evaluando la columna para decidir el tratamiento de datos, nos damos cuenta que: los datos faltantes corresponden a un 17,165%  (1463   ÷   8523) lo que implica que no es un % significativo, por lo que procederemos a reemplazar los valores faltantes con la moda.

```python
# reemplazamos con la moda.
df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
```

```python
# revisamos si es que siguen existiendo valores nulos en 'Item_Weight'
df.isna().sum()
```
